```python
import pandas as pd
import numpy as np
import random 
```

```python
cust1 = pd.read_csv('/Users/kalindjoshi/Desktop/ass3/Customer_Names.csv')
# reading the file from directory
```

```python
cust1.isnull().any()
# checking for any null values
```

```python
cust1 = cust1.drop(['Unnamed: 2'	,'Unnamed: 3'	,'Unnamed: 4'	,'Unnamed: 5'	,'Unnamed: 6',	'Unnamed: 7'	,'Unnamed: 8',	'Unnamed: 9'], axis=1)
#dropping unwanted columns
```

```python
cust1.head()
```

```python
cust1.columns = map(str.lower, cust1.columns)

cust1['first name'] = cust1['first name'].str.lower()
cust1['last name'] = cust1['last name'].str.lower()

#converting column titles to lower case
```

```python
cust1.head()
```

```python
cust1["email"] = cust1["first name"] + cust1["last name"]

# creating new column email combination of first and last name
```

```python
cust1.head()
```

```python
cust1['custno'] = np.random.randint(1000000000, 9999999999, cust1.shape[0])

# generating numbers in range
```

```python
cust1.head()
```

```python
start_u = (pd.to_datetime('2000-01-01')).value//10**9
end_u = (pd.to_datetime('2022-11-20')).value//10**9

cust1['custjoining'] = pd.to_datetime(np.random.randint(start_u, end_u, cust1.shape[0]), unit='s')

# generating dates 
```

```python
cust1.head()
```

```python
cust1['custzip'] = np.random.randint(10000, 99999, cust1.shape[0])

#generating zip codes
```

```python
cust1.head()
```

```python
cust1.reset_index(inplace = True)
cust1 = cust1.rename(columns={'index':'custid'})
```
