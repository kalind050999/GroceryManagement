```python
prod1 = prod1.rename(columns={'Member_number': 'prodid', 'itemDescription': 'prodname'})
#renaming the column names according to my table column names
```

```python
prod1.head()
```

```python
prod1.drop('Date',axis=1)
# dropping date column
```

```python
prod1['proddis'] = np.random.randint(0, 10, prod1.shape[0])
```

```python
prod1['prodmp'] = np.random.randint(1, 17, prod1.shape[0])
prod1['prodsp'] = np.random.randint(1, 17, prod1.shape[0])
# generating random values for selling and market price in a range
```

```python
prod1['prodprice'] = prod1['prodsp'] - (prod1['prodsp']*(prod1['proddis']/100))

#calculating product selling price by formula
```

```python
prod1 = prod1.drop('Date', axis=1)
```

```python
prod1['prodcat'] = ''
prod1.head()
```
