```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
prodeda = pd.read_csv("/Users/kalindjoshi/Desktop/ass3/prodclean.csv")

#opening the file
```

```python
prodeda.prodcat.value_counts()

#getting count of all rows which repeat
```

```python
prodeda.prodcat.value_counts().plot(kind="bar")
plt.title("")
plt.xlabel("")
plt.xticks(rotation=90)
plt.ylabel("")
plt.show()

#bar graph between price and product
```

```python
sns.catplot(x="prodprice", y="prodcat", data=prodeda, kind="box", aspect=1.5)
plt.title("")
plt.show()

#box plot between category and price
```

```python
sns.scatterplot(x="prodcat", y="prodprice", hue="proddis", data=prodeda, palette="Dark2", s=80)
plt.title("")
plt.xticks(rotation=90)
plt.show()

#scatterplot with price, discount and category
```

```python
corrmat = prodeda.corr()
hm = sns.heatmap(corrmat, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=prodeda.columns, 
                 xticklabels=prodeda.columns, 
                 cmap="Spectral_r")
plt.show()

#heat map correlation
```
